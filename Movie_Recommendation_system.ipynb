# Data Preprocessing

```python
type(data.iloc[0, 3])
```

```python
import ast
def convert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i["name"])
    return l
```

```python
data["genres"] = data["genres"].apply(convert)
data["keywords"] = data["keywords"].apply(convert)
```

```python
def converter(obj):
    l = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            l.append(i["name"])
            counter = counter + 1
        else:
            break
    return l
```

```python
data["cast"] = data["cast"].apply(converter)
```

```python
import ast
def fetch_director(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i["job"] == "Director":
            l.append(i["name"])
            break
    return l
```

```python
data["crew"] = data["crew"].apply(fetch_director)
```

```python
data.isnull().sum()
```

```python
data.sample(3)
```

```python
data["overview"]= data["overview"].fillna("") .apply(lambda x : x.split())
```

```python
data.sample(2)
```

```python
data["genres"] = data["genres"].apply(lambda x: [i.replace(" ", "") for i in x])
data["keywords"] = data["keywords"].apply(lambda x: [i.replace(" ", "") for i in x])
data["cast"] = data["cast"].apply(lambda x: [i.replace(" ", "") for i in x])
data["crew"] = data["crew"].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
data["tag"] = data["overview"] + data["genres"] + data["keywords"] + data["cast"] + data["crew"]
```

```python
data.sample(2)
```

```python
p = ["overview",	"genres",	"keywords", "cast", "crew"]
for i in p:
    data.drop(columns = i, axis = 1, inplace = True)
```

```python
data
```

```python
data["tag"] = data["tag"].apply(lambda x : " ".join(x))
```

```python
data.sample(2)
```

```python
data["tag"] = data["tag"].apply(lambda x : x.lower())
```

```python
data.sample(2)
```

```python
from nltk.stem import PorterStemmer
p = PorterStemmer()
v = []
def stem(obj):
    for i in obj.split():
        v.append(p.stem(i))
    return " ".join(v)
```

```python
data["tag"] = data["tag"].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
c = CountVectorizer(max_features=5000, stop_words="english")
```

```python
vector = c.fit_transform(data["tag"]).toarray()
```

```python
vector.shape
```

```python
d = c.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
result = cosine_similarity(vector)
```
